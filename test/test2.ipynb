```python
import cv2
import numpy as np
from sklearn.mixture import GaussianMixture as GMM

img = cv2.imdecode(np.fromfile("./test.jpg", dtype=np.uint8), -1)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

ch1 = img_rgb[:, :, 0].flatten()
ch2 = img_rgb[:, :, 1].flatten()
ch3 = img_rgb[:, :, 2].flatten()

rawlist = np.vstack((ch1, ch2, ch3))
rawlist = rawlist.astype(np.double)
```

```python
def find_in_list(lst, k):
    index_lst = []
    for i in range(len(lst)):
        if lst[i] == k:
            index_lst.append(i)
    return index_lst
```

```python
gmm = GMM(n_components=4, max_iter=1000).fit(rawlist.T)
clusters = gmm.predict(rawlist.T)
labels_list = list(clusters)

rawcolorlist = np.array([])
```
